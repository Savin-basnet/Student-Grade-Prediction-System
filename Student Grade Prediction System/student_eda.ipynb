```python
data = data.drop(columns=['Unnamed: 0'])
```

```python
# Replacing median values for age, Medu, Fedu nan values 

data['age'] = data['age'].fillna(data['age'].median())
data['Medu'] = data['Medu'].fillna(data['Medu'].median())
data['Fedu'] = data['Fedu'].fillna(data['Fedu'].median())
```

```python
# data.head(6)
```

```python
# using mode imputation of missing values

data['Pstatus'] = data['Pstatus'].fillna(data['Pstatus'].mode()[0])
data['reason'] = data['reason'].fillna(data['reason'].mode()[0])
data['internet'] = data['internet'].fillna(data['internet'].mode()[0])
```

```python
# data.head()
```

```python
# data.isna().sum()
```

```python
# identifying categorical columns in the dataset

categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
print(categorical_columns)
```

```python
# handling binary columns if it has "yes" then 0 else it has "No" then it converts to 1

binary_cols = ['schoolsup', 'famsup', 'paid', 'activities', 
               'nursery', 'higher', 'internet', 'romantic']
import numpy as np
for col in binary_cols:
    data[col] = np.where(data[col] == 'yes', 1, 0)
```

```python
# data.head()
```

```python
# data.columns
```

```python
non_binary_cats = ['school', 'sex', 'address', 'famsize', 'Pstatus', 
                   'Mjob', 'Fjob', 'reason', 'guardian'] # female becomes 0 and male becomes 1 in sex_M column
data = pd.get_dummies(data, columns=non_binary_cats, drop_first=True) #here drop first makes the first as 0 and second as 1 
```

```python
#converting boolean values to integer for cleaner interpretation
# Select boolean columns

bool_cols = data.select_dtypes(include=['bool']).columns

# Convert to int
data[bool_cols] = data[bool_cols].astype(int)
```

```python
#converting float values to integer for cleaner interpretation

# Convert back to integer (rounding just in case)
data['Medu'] = data['Medu'].round().astype(int)
data['Fedu'] = data['Fedu'].round().astype(int)
data['age'] = data['age'].round().astype(int)   # age is whole number anyway
```
